```python
import duckdb
import plotnine as pn
import numpy as np
```

```python
db = duckdb.connect("results.db")
```

```python
df = db.sql("SELECT * FROM stats").df()
```

```python
models = [
    "normal",
     "ill-normal",
     "corr-normal",
    # "rosenbrock",
    "glmm-poisson",
    "hmm",
    "garch",
    # "lotka-volterra",
     "arma",
    # "arK"
]
idx = df["model"].isin(models)
df = df.loc[idx, :]
```

```python
pn.ggplot(df) + \
    pn.geom_jitter(pn.aes("algorithm", "msjd", color = "algorithm"), alpha = 0.5) + \
    pn.geom_boxplot(pn.aes("algorithm", "msjd", color = "algorithm"), fill = None) + \
    pn.facet_wrap("~model", scales = "free", nrow = 1) + \
    pn.theme_minimal()
```
