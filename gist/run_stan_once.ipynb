```python
import cmdstanpy as csp
import numpy as np
import pandas as pd
```

```python
cfg = {
    "model_path": "../stan",
    "model_name": "glmm-poisson"
}
```

```python
stan_file = f"{cfg['model_path']}/{cfg['model_name']}.stan"
data_file = f"{cfg['model_path']}/{cfg['model_name']}.json"
stan_model = csp.CmdStanModel(stan_file = stan_file, force_compile = True)
```

```python
fit = stan_model.sample(data = data_file,
                        iter_warmup = 25_000,
                        iter_sampling = 50_000,
                        show_progress = False,
                        show_console = False)
```

```python
fit.summary()
```

```python
df = fit.draws_pd()
```
